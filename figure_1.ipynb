#### Stolper-Samuelson: 
increase in the price of a capital intensive product will increase real incomes of owners of capital and  
decrease real incomes of owners of the other factors, namely the labour owners workers' income wages

```python
# Load the FRED API KEY
import os
from dotenv import load_dotenv

load_dotenv()
api_key = os.environ.get("FRED_API_KEY")
```

```python
# Load US wages data
from fredapi import Fred
fred = Fred(api_key=api_key)
wages = fred.get_series('LES1252881600Q')
```

```python
# Plotting the wages
import pandas as pd
import matplotlib.pyplot as plt

# Define the style
plt.style.use("ggplot")

# Plot the wages time series
fig = plt.figure(figsize=(12, 6))
plt.plot(wages)

# Adding event for context
event_date_1 = pd.to_datetime('2001-12-11') # China in WTO
event_date_2 = pd.to_datetime('2017-01-20') # Trump inagurated

plt.axvline(
    x=event_date_1,             
    color='blue', 
    linestyle='--', 
    linewidth=1.5,
    label='Event 1'
)

plt.axvline(
    x=event_date_2,            
    color='blue', 
    linestyle='--', 
    linewidth=1.5,
    label='Event 2'
)

fig.text(
    x=0.51, 
    y=0.75,         # Y position slightly above the mean for visibility
    s='China in WTO',                # The text string
    color='blue', 
    fontsize=10, 
    fontweight='bold',
    rotation=90,                # Rotate text 90 degrees vertically
    ha='right',                 # Align text to the right of the vertical line
    va='center'
)

fig.text(
    x=0.79, 
    y=0.75,         # Y position slightly above the mean for visibility
    s='Trump president',                # The text string
    color='blue', 
    fontsize=10, 
    fontweight='bold',
    rotation=90,                # Rotate text 90 degrees vertically
    ha='right',                 # Align text to the right of the vertical line
    va='center'
)

plt.ylabel("Real Earning in USD")
plt.title("Median usual weekly real earnings")

fig.text(
    0.02,                   # X-coordinate (0.0 = far left, 1.0 = far right)
    0.01,                   # Y-coordinate (0.0 = bottom, 1.0 = top)
    "Source: FRED", 
    fontsize=8,             # Keep it small
    color='gray',
    ha='left'               # Horizontal alignment
)

fig.tight_layout(rect=[0, 0.03, 1, 1])

# plt.axvline("2001-12-11", color='red', linestyle='--')

plt.show()
```
