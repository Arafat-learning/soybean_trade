```python
import numpy as np

# data from https://apps.fas.usda.gov/gats/ExpressQuery1.aspx
world = np.array([18.861872, 22.839176, 21.456288, 17.058077, 18.693613, 25.516208, 27.418050, 34.352542])
china = np.array([10.489267, 14.203109,	12.224950, 3.120091, 8.005228, 14.068113, 14.116249, 17.917513])
```

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")

fig = plt.figure(figsize=(12, 6))

plt.bar(np.arange(2015, 2023), world, label = "World")
plt.bar(np.arange(2015, 2023), china, label = "China")

plt.title("US soybean exports")
plt.ylabel("Export (billion USD)")
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
fig.text(
    0.02,                   # X-coordinate (0.0 = far left, 1.0 = far right)
    0.01,                   # Y-coordinate (0.0 = bottom, 1.0 = top)
    "Source: USDA FAS", 
    fontsize=8,             # Keep it small
    color='gray',
    ha='left'               # Horizontal alignment
)
fig.tight_layout(rect=[0, 0.03, 1, 1])
plt.show()
```
