```python
import numpy as np

q = np.arange(100)
q = q/10
d = 10-q
s = q
no_tax = np.full(100, 2.5)
tax = np.full(100,4.5)
```

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")

fig = plt.figure(figsize=(12, 6))

plt.plot(q, d)
plt.plot(q, s)
plt.plot(q, no_tax) 
plt.plot(q, tax)

fig.text(
    0.15,                   # X-coordinate (0.0 = far left, 1.0 = far right)
    0.485,                   # Y-coordinate (0.0 = bottom, 1.0 = top)
    "US export with tariff", 
    fontsize=9,             # Keep it small
    color='black',
    ha='left'               # Horizontal alignment
)
fig.text(
    0.15,                   # X-coordinate (0.0 = far left, 1.0 = far right)
    0.335,                   # Y-coordinate (0.0 = bottom, 1.0 = top)
    "US export without tariff", 
    fontsize=9,             # Keep it small
    color='black',
    ha='left'               # Horizontal alignment
)

plt.title("")
plt.ylabel("Price")
plt.xlabel("Quantity")
plt.ticklabel_format(style='plain', axis='y')
fig.tight_layout(rect=[0, 0.03, 1, 1])
plt.show()

plt.show()
```
